Utilities -- typically minor decorators
    alliterate()          mark the alliteratitive potential on all words in an annotated document before lineation
    numberlines(start)    number the <vg> elements in lineated document - will become part of lineation
    decoratelines()       add common (empty) attributes to <vg> elements - will become part of lineation

Transformers -- these will restructure the document
    syllabify()           break the words <w> in a lineated document into individual syllables <s>. 
    importscan()          turn a scan file into metrical markup to produce an analyzed document

Reversion transformers
    de_analyze()          revert an analyzed to a syllabified document
    de_syllabify()        revert a syllabified to a lineated document
    de_lineate()          revert a lineated to an annotated document
